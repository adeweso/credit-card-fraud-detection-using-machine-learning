# HANDLING DATA IMBALANCE - OVERSAMPLING

```python
data
```

```python
data["Class"].value_counts()
```

```python
data.duplicated().any()
```

```python
x = data.drop("Class", axis = 1)   # feature matrix (all columns except "Class")
y = data["Class"]  # target variable (only the "Class" column)
```

```python
x
```

```python
y
```

```python
x.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
x_res, y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
y_res
```

```python
x_res
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_res, y_res, test_size = 0.20, random_state = 42)
# 20% for testing, 80% for training. # random_state = 42: Ensures the split is reproducible (same results every time).
```

# LOGISTIC REGRESSION - AFTER OVERSAMPLING

```python
log = LogisticRegression()
log.fit(x_train, y_train)    # trains the model on the training sets x_train(features) and y_train(labels)
```

```python
y_pred1 = log.predict(x_test)  # x_test used to evaluate model performance   # it makes predictions(labels) using the trained model
```

```python
accuracy_score(y_test, y_pred1)   # y_test used to check if the predictions are correct
```

```python
precision_score(y_test, y_pred1)
```

```python
recall_score(y_test, y_pred1)
```

```python
f1_score(y_test, y_pred1)
```

# DECISION TREE - OVERSAMPLING

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
 y_pred2 = dt.predict(x_test)  # x_test used to evaluate model performance   # it makes predictions(labels) using the trained model
```

```python
accuracy_score(y_test, y_pred2)
```

```python
precision_score(y_test, y_pred2)
```

```python
recall_score(y_test, y_pred2)
```

```python
f1_score(y_test, y_pred2)
```

# RANDOM FOREST AFTER OVERSAMPLING

```python
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
y_pred3 = rf.predict(x_test)
```

```python
accuracy_score(y_test, y_pred3)
```

```python
precision_score(y_test, y_pred3)
```
