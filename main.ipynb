```python
import tkinter as tk
import customtkinter as ctk
from tkinter import messagebox, ttk
import joblib

def show_entry_fields():
    global result_label

    try:
        values = [float(entry.get()) for entry in entries]
    except ValueError:
        result_label.configure(text="Please enter valid numeric values.")
        return

    try:
        model = joblib.load('credit card model')  # make sure the model file exists
        y_pred = model.predict([values])
        result = "Normal Transaction" if y_pred[0] == 0 else "Fraudulent Transaction"
        print("Credit Card Fraud Detection System:", result)
        result_label.configure(text=f"Prediction: {result}")
    except Exception as e:
        result_label.configure(text=f"Error: {e}")

def initialize_homepage():
    global entries, result_label

    entries = []
    for i in range(29):
        label = ctk.CTkLabel(app, text=f"Enter value of V{i+1}", font=("Arial", 14), text_color="black")
        label.grid(row=i % 10 + 1, column=(i // 10) * 2, padx=20, pady=10, sticky="w")

        entry = ctk.CTkEntry(app, width=250, height=40, placeholder_text="Enter value",
                             font=("Arial", 16), corner_radius=10, border_width=2,
                             fg_color="white", text_color="black")
        entry.grid(row=i % 10 + 1, column=(i // 10) * 2 + 1, padx=10, pady=5)
        entries.append(entry)

    predict_button = ctk.CTkButton(app, text="Predict", font=("Arial", 14, "bold"),
                                   command=show_entry_fields, width=200, height=40,
                                   corner_radius=10, fg_color="blue", text_color="white")
    predict_button.grid(row=11, column=0, columnspan=6, pady=20)

    result_label = ctk.CTkLabel(app, text="Prediction: ", font=("Segoe UI", 18), text_color="black")
    result_label.place(x=500, y=600)  # Adjust the x and y values as needed


# Initialize the application
app = ctk.CTk()
app.geometry("1200x600")
app.configure(fg_color='white')
app.title("Credit Card Fraud Detection System")

# Center the window
window_width = 1300
window_height = 650
screen_width = app.winfo_screenwidth()
screen_height = app.winfo_screenheight()
x_cordinate = int((screen_width / 2) - (window_width / 2))
y_cordinate = int((screen_height / 2) - (window_height / 2))
app.geometry(f"{window_width}x{window_height}+{x_cordinate}+{y_cordinate}")

initialize_homepage()
app.mainloop()
```
